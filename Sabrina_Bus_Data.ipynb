```python
base_url = 'https://geoservices.tamu.edu/Services/ReverseGeocoding/WebService/v04_01/Rest/?'

version = 4.10
query_url = f'{base_url}apiKey={t_key}&version=4.10&format=json&state=az&notStore=false'

address = []
city = []
zipcode = []

for index,row in clean_bus_df.iterrows():
    query_api = query_url+'&lat='+str(row['Y'])+'&lon='+str(row['X'])
    print(query_api)
    response = requests.get(query_api).json()
    print(response)
    zipcode.append(response["StreetAddresses"][0]['Zip'])
    city.append(response["StreetAddresses"][0]['City'])
    address.append(response["StreetAddresses"][0]['StreetAddress'])
```
