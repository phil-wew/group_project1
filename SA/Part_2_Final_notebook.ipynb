```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
import json
import urllib, json
import scipy.stats as st

from scipy.stats import linregress
from census import Census
from config import api_key
c = Census(api_key, year=2018)
```

```python
#Read in 4 CSV's that contain the data and the zipcodes associated with each stop
new_csv = "../CSVs/filter_bus_1000.csv"
new_df = pd.read_csv(new_csv,"utf-8", delimiter = ",")

two_csv = "../CSVs/new_bus_2.csv"
two_df = pd.read_csv(two_csv,"utf-8", delimiter = ",")

three_csv = "../CSVs/filter_bus_3001-4500.csv"
three_df = pd.read_csv(three_csv,"utf-8", delimiter = ",")

four_csv = "../CSVs/filter_bus_4500-5535.csv"
four_df = pd.read_csv(four_csv,"utf-8", delimiter = ",")
```

```python
#Merge all CSV's together to create one dataeframe
updated_df = pd.concat([new_df, two_df, three_df, four_df])
#Display dataframe 
updated_df
```

```python
#Deleted extra columns
final_df = updated_df.drop(['WaterFountain','ADA','ComputedState','X', 'Y', 'ComputedZipPlus4','TransactionId','Column1', 'Zipcode', 'StreetAddress', 'AUTO_UNIQUE_ID_2020-07-28_smavila1_new_bus_2','OBJECTID','AUTO_UNIQUE_ID_2020-07-28_czc5230_bus_4','AUTO_UNIQUE_ID_2020-07-27_cceni03_filter_bus3001', 'DateUpdated','Routes', 'Zip', 'Type_', 'Direction','AUTO_UNIQUE_ID_2020-07-27_pwewiora08_filter_bus_1000','NoName','City', 'Address', 'TimeTaken','Source','ComputedAPN', 'State', 'UpdatedReverseGeocoding', 'ErrorMessage','ATIS_ID','NextRide', 'Version'], axis=1)
#Dropped duplicates in case there was overlap
final_df.drop_duplicates(keep=False,inplace=True)
final_df.rename(columns = {'ComputedZip':'Zipcode', 'ComputedCity':'City', 
                              'ComputedStreetAddress':'Street Address'}, inplace = True) 
#Display updated datafram
final_df
```

```python
#Request specific data from the census
census_data = c.acs5.get(('B19301_001E', 
                          'B01003_001E',"B17001_002E"), {'for': 'zip code tabulation area:*'})

#Variable type
#B19301_001E = Per Capita Income
#B01003_001E = total population
#"B17001_002E" = poverty count
#Create census dataframe
census_pd = pd.DataFrame(census_data)
census_pd = census_pd.rename(columns={"B19301_001E" : "Per Capita Income",
                                      "B17001_002E" : "Poverty Count",
                                      "B01003_001E" : "Population",
                                      "zip code tabulation area": "Zipcode"})
census_pd = census_pd.fillna(0)

# Add in Poverty Rate (Poverty Count / Population)
census_pd["Poverty Rate"] = 100 * \
    census_pd["Poverty Count"].astype(
        int) / census_pd["Population"].astype(int)
census_pd['Zipcode'] = census_pd['Zipcode'].transform(float)
```

```python
#Merge census data and bus data togther on zipcode 
bus_census = pd.merge(final_df, census_pd, how='left', on=['Zipcode'])
```

```python
#Create datafrmae with only "Jusridication" Phoenix in order to compare to other jurisdictions 
phx_bus = bus_census.loc[bus_census["Juris"]=="Phoenix"]
#Display dataframe
phx_bus
```

```python
#Getting a count for each bus stop (row) in each zipcode
bus_counts = phx_bus['Zipcode'].value_counts(ascending=True)
#Displaying count
bus_counts
```

```python
#Creating a bar graph to show how many bus stops are in each zipcode
plt.figure(figsize=(15,15))
bus_counts.plot(kind = 'bar')
#Creating axis labels and graph title
plt.xlabel("Zipcode")
plt.ylabel("Number of Bus Stops")
plt.title("Bus Stops vs. Zipcodes")
#Saving image as png
plt.savefig("zip_bus.png")
```

```python
#Create datafrmae with only "Jusridication" Tempe in order to compare to other jurisdictions 
tempe_bus = bus_census.loc[bus_census["Juris"]=="Tempe"]
```

```python
#Create datafrmae with only "Jusridication" Mesa in order to compare to other jurisdictions 
mesa_bus = bus_census.loc[bus_census["Juris"]=="Mesa"]
```

```python
#Grouping by zipcode to get the average's for each census variable
pov_r = phx_bus.groupby("Zipcode").mean()["Poverty Rate"]
pov_c = phx_bus.groupby("Zipcode").mean()["Poverty Count"]
pop = phx_bus.groupby("Zipcode").mean()["Population"]
per_cap = phx_bus.groupby("Zipcode").mean()["Per Capita Income"]
#Creating new dataframe to display
new_bus = pd.DataFrame({"Per Capita Income": per_cap,
                        "Population": pop,
                        "Poverty Count": pov_c,
                        "Poverty Rate": pov_r,
                       "Number of Bus Stops": bus_counts})
#Setting the index as zipcode
new_bus.index.name = 'Zipcode'
#Displaying dataframe
new_bus
```
