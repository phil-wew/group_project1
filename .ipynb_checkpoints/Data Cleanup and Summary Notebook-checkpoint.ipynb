**data cleaning**: We found datasets provided through the Valley Metro developers page that contained data for every bus stop, rail stop, ridership numbers spanning over a number of years and broken down by quarter. We began with the csv file that had every bus and rail stop listed. 

This file contained latitude and longitude for every station, yet did not provide zipcodes. Our group was planning to use the census API. So, in order to match census data with where bus stops are located we need to do reverse geocoding with the latitude and longitude to retrieve zipcodes.

Google offers a reverse geocoding service, but no one wanted to risk being charged for exceeding our API credit limit. We used the Texas A&M free service. However, we still ran into problem of hitting the free limit with the csv from Valley Metro. The csv had a very large set of data.

We split the original csv into four parts in order to each run one and remain within the API credit limit. Texas A&M returned back our csv with zipcodes and more completed addresses. At this point we did a simple `.concat()` to bring the four parts back together as one.

At this point we had a complete list of stations with the address and zipcode. We then merged this dataset with the csv containing ridership numbers by quarter and year resulting in *all_bus_riders.csv.*

After this we had to run the census API. We pulled the 5 year American Community Survey Data by zipcode. Of course, this gave every zipcode for the country, so we needed to clean this up a bit as well. We took the new df of census data and merged it with our all_bus_riders on the zipcode in order to just pull over the information pertaining to Valley Metro zipcodes.

The next step was grouping by Zipcode and finding the sum and mean of most of our columns, such as the total number of bus stops in each zipcode, or the mean per capita income for a zipcode. The bus data was also split into weekday and weekend sets in order to determine if there are seperate trends depending the part of the week.


**Question 2.5: How does income and poverty affect public transit use? Are there obvious trends in ridership numbers and the poverty/income/population numbers of a given zipcode?**

What we generally found was that percent of poverty, income and even population of a zipcode did not have a strong relationship, and cannot be used a predictor for ridership numbers of Valley Metro busses in Phoenix. See Figures 1-6. Population and then poverty rate had the greatest relationship to ridership numbers on weekdays, however that relationship was still very weak.
