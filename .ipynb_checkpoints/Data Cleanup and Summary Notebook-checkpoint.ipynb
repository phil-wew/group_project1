# The Mission

With everyone being stuck at home due to the pandemic, we wanted to see the effects that it was having on public transportaion in Arizona. After looking at the Valley Metro CSV's we decided that there were too many data points and that we needed to narrow down our search. We agreed to look at only the cities where the lightrail was in order to do a comparison of bus stops to rail stops which gave us three cities; Phoenix, Tempe and Mesa. Our mission was to find the anwsers to the following questions for each of the three cities that we wanted to focus on, our questions were: 
* How has COVID impacted bus/rail ridership? 
* How does income and poverty affect public transit use? 
* Are there obvious trends in ridership numbers and the poverty/income/population numbers of a given zipcode? 
* What kind of effect does popluation have on ridership?

Our expectations are the ridership will be effected by the population in that higher zipcode populations will give us a reuslt of more ridership. Lower income zipcodes will have more bus stops, and that there will be obvious trends in ridership comparing weekdays to weekends. We also assume that COVID has drastically decreased the amount of ridership on both bus stops and rail riding. 

**Data cleaning**: We found datasets provided through the Valley Metro developers page that contained data for every bus stop, rail stop, ridership numbers spanning over a number of years and broken down by quarter. However, there was no recent data in order for us to look at COVID. We had to reach out to Valley Metro to ask for updated data and we were given this data for the rail system only. 

We began with the csv file that had every bus and rail stop listed. This file contained latitude and longitude for every stop and station, yet did not provide zipcodes. Since we were planning to use the census API we had to do reverse geocoding with the latitude and longitude to retrieve zipcodes. This was in order to match census data with where bus and rail stops are located in the jursidictions. 

Google offers a reverse geocoding service, but no one wanted to risk being charged for exceeding our API credit limit. We used the Texas A&M free service. However, we still ran into problem of hitting the free limit with the csv from Valley Metro. The csv had a very large set of data (over 5500 rows).

We split the original csv into four parts in order to each run one and remain within the API credit limit. Texas A&M returned back our csv with zipcodes and more completed addresses. At this point we did a simple `.concat()` to bring the four parts (*filter_bus_1000, filter_bus_1000-3001, filter_bus_3001-4500, filter_bus_4500-5535*) back together as one.

At this point we had a complete list of stations with the address and zipcode. We then merged this dataset with the csv containing ridership numbers by quarter and year resulting in *all_bus_riders.csv.*

After this we had to run the census API. We pulled the 5 year American Community Survey Data by zipcode. Of course, this gave every zipcode for the country, so we needed to clean this up a bit as well. We took the new df of census data and merged it with our all_bus_riders on the zipcode in order to just pull over the information pertaining to Valley Metro zipcodes.

The next step was grouping by Zipcode and finding the sum and mean of most of our columns, such as the total number of bus stops in each zipcode, or the mean per capita income for a zipcode. The bus data was also split into weekday and weekend sets in order to determine if there are seperate trends depending the part of the week.

By this time we had only a couple of days before the presentation and we realized that we had to change our scope and narrow down which data to look at and have questions that better fit the data once it was cleaned. We decided to narrow down our scope to only Phoenix.

# How does population effect bus ridership in the Phoenix area?

This portion focused on ridership vs population in Phoenix, Tempe, and Mesa. The ridership was greatest in the Phoenix area followed by Tempe. As the routes moved further away from the central part of Phoenix, the ridership dipped significantly with Mesa having a low figure compared to the population. This is probably due to the need for a vehicle to travel further distances in a reasonable time especially for people who live in the East County and commute into or around the Phoenix city limits.



# How does population effect rail ridership in the Phoenix area?

The ridership for the rail line continues to grow as more stops are added into the East Valley areas and is becoming more of an option for commuters. However, the numbers are still lower than what it could be as more people are still learning the rail lines and how to use them properly to be efficient in traveling to their destinations. 



# How has Covid-19 impacted rail ridership?

From the Analysis conducted, we determined that COVID has had a significant impact on ridership in three major cities in Arizona. Tempe is the least impacted city; ridership has decreased significantly from 2019 with its most impacted month being in April which was when most of the country went on lockdown. It is evident from the data that ridership has increased over the last few months. 



Mesa saw promising increases in January and February 2020. It then took a hit in the number of ridership in March and April plummeting all the way down to -113% (compared to those from last year) when COVID struck. Ridership slowly started increasing back up in May but took a slight hit in June. 



Phoenix is the city that was the most impacted by COVID, ridership slightly decrease in January due to not many people riding after holiday season. Ridership increase in February then took a big hit in March and April when COVID hit. Ridership slightly increased from -161% in April to -104% in May & -103% in June and has remained relatively the same since, however, ridership has been impacted the worst in Phoenix when comparing to 2019 totals.
