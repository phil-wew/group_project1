```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
from scipy.stats import linregress

from config import t_key 
```

```python
bus_csv = "CSVs/Valley_Metro_Bus_Stops_View.csv"

bus_stops_df = pd.read_csv(bus_csv, encoding="utf-8")
bus_stops_df.head()
```

```python
bus_df = bus_stops_df.drop(['NextRide', 'Photo', 'ATIS_ID', 'Em_Call', 'Lighting','OBJECTID', 'Routes', 'Bathroom','Direction', 'DateUpdated', 'ADA', 'PublicArt', 'Shelter', 'Bench','BusBay', 'BikeRack', 'WaterFountain', 'TrashCan', 'Info_Display', 'Sidewalk'], axis=1)
bus_df.head()
```

```python
new_bus_df = bus_df.copy()

new_bus_df.set_index("Juris", inplace = True)

final_df = new_bus_df.loc[['Mesa','Tempe','Phoenix']]
final_df 
```

```python

bus_df =final_df.reset_index()

bus_df["Zipcode"] = " "
bus_df["City"] = " "
bus_df["Street Address"] = " "

#print(bus_df)

df_one = bus_df.iloc[0: 1000]

df_two = bus_df.iloc[1000: 2766]

df_three = bus_df.iloc[2766: 4150]

df_four = bus_df.iloc[4150: 5535]



df_one.to_csv('bus_1_CSV')
df_two.to_csv('bus_2_CSV')
df_three.to_csv('bus_3_CSV')
df_four.to_csv('bus_4_CSV')

```

```python
base_url = 'https://geoservices.tamu.edu/Services/ReverseGeocoding/WebService/v04_01/Rest/?'

version = 4.10
query_url = f'{base_url}apiKey={t_key}&version=4.10&format=json&state=az&notStore=false'

address = []
city = []
zipcode = []

for index,row in final_df.iterrows():
    query_api = query_url+'&lat='+str(row['Y'])+'&lon='+str(row['X'])
    print(query_api)
    response = requests.get(query_api).json()
    zipcode.append(response["StreetAddresses"][0]['Zip'])
```
